**Importing Libraries** üìö

```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

**Getting unique list of Artists Spotify id and name** üé§

```python
all_artists_df = pd.read_csv("old_artists.csv")
all_artists_df = all_artists_df[["id", "name"]]
all_artists_df.drop_duplicates(inplace = True)
all_artists_df
```

**Finding list of artists with only id and not name** üîé

```python
selected_artists_df = pd.read_csv("E:/Quera/Spotify_Project/Spotify_Data/track_info.csv")
selected_artists_list = selected_artists_df["artist_id"].drop_duplicates().tolist()
all_artists_list = all_artists_df["id"].drop_duplicates().tolist()

difference_list = [artist_id for artist_id in selected_artists_list if artist_id not in all_artists_list]
```

**Getting Artists Name** üí≥

```python
new_artist_name = []

def get_artist_name(search_term):
    chrome_options = Options()
    chrome_options.add_argument("--lang=en-US")
    driver = webdriver.Chrome(options=chrome_options)
    driver.get(f"https://www.google.com/search?q={search_term}")
    
    try:
        artist_element = driver.find_element("class name", "LC20lb")
        artist_name = artist_element.text
    except Exception as e:
        artist_name = None
    
    driver.quit()
    return artist_name

for i in difference_list:
    search_term = i
    artist_name = get_artist_name(search_term)
    new_artist_name += [artist_name]
```

```python
data_dict = {"artist_id": difference_list, "artist_name": new_artist_name}
df = pd.DataFrame(data_dict)
```

**Gathering all artists data together** üñá

```python
selected_artists_df = selected_artists_df[["artist_id"]]
selected_artists_df = selected_artists_df.drop_duplicates()
selected_artists_df
```

```python
selected_artists_df = selected_artists_df.merge(all_artists_df, left_on = "artist_id", right_on = "id", how = "left")
selected_artists_df.drop(columns = ["id"], inplace = True)
selected_artists_df.rename(columns = {"name": "artist_name"}, inplace = True)
selected_artists_df
```

```python
output_df = df
final_artists_df = selected_artists_df
final_artists_df.dropna(subset = ["artist_name"], inplace = True)
final_artists_df
```

```python
combined_df = pd.concat([final_artists_df, output_df])
combined_df.to_csv("artists.csv", index = False)
```
